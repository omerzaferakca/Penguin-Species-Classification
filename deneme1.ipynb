```python
# Gerekli kütüphanelerin içe aktarılması
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data okuma işlemi
data = pd.read_csv('penguins_data.csv')
```

```python
# Verinin ilk 5 satırını okuma, isteğe bağlı içerisine girilen değer kadar satır getirir
data.head()
```

```python
# Verinin son 5 satırını okuma, isteğe bağlı içerisine girilen değer kadar satır getirir
data.tail()
```

```python
# Veri seti hakkında  bilgi
data.info()
```

```python
# Veri setinin satır  ve sütun sayısı
data.shape
```

```python
# Sex sütunundaki '.' anlamsız verisini MALE olarak değiştirdik
data['Sex'] = data['Sex'].replace('.', 'MALE')
```

```python
# Sex sütunundaki Female ve Male değerlerinin F ve M olarak değiştirilmesi
data['Sex'] = data['Sex'].replace(['FEMALE', 'MALE'], ['F', 'M'])
```

```python
# Comments sütununun kaldırılması, inplace seçeneği veri setine uygulanması için gerekli
data.drop(labels=['Comments'], axis=1, inplace=True)
```

```python
data.drop(labels=['studyName'], axis=1, inplace=True)
```

```python
data.head()
```

```python
# Groupby metodu kullanılarak türlerin cinsiyetine göre diğer sütunlardaki değerlerini gösterme 
data.groupby(by=['Species','Sex']).count()
```

```python
# Species kolonunun sayısını gösteren komut
data['Species'].value_counts()
```

```python
data['Island'].value_counts()
```

```python
# Sütunlar arasındaki korelasyonu belirtir, numeric only komutu ile sadece sayısal değeri olan sütunlar alınır
data_corr = data.corr(numeric_only=True)
```

```python
# Korelasyonu grafik üzerinde gösterdik. Annot sayısal değerlerin gösterilmesini, cmap temayı belirtir.
sns.heatmap(data_corr, annot=True, cmap='coolwarm')
```

```python
# Veri setindeki boş değerlerin toplamını, verilerin sayısına bölerek yüzdesini alıyoruz
nan_percentage = data.isna().sum() / data.count() * 100

# Boş verilerin toplam sayısını belirledik
nan_count = data.isna().sum()

# Boş verilerin toplamını ve yüzdelik halini tablo haline getirdik. Axis = 0 satırlar için, Axis = 1 sütunlar için
nan_table = pd.concat([nan_count, nan_percentage], axis=1)

# Tablonun sütunlarının isimlerini belirledik
nan_table.columns = ['Count', 'Percentage']
nan_table
```

```python
# Boş verileri doldurmak için SimpleImputer sınıfını çağırdık
from sklearn.impute import SimpleImputer
```

```python
# Doldururken en çok kullanılan değeri alıyoruz
imputer = SimpleImputer(strategy='most_frequent')
```

```python
# iloc[:, :] ile işlemi veri setinin tüm değerleri için yapmak istediğimizi belirttik. Fit_transform metoduyla veri setine eksik değerlerin doldurulmasını ve dönüştürülmesini sağladık.
data.iloc[:, :] = imputer.fit_transform(data)
```

```python
data.isna().sum()
```

```python
# Cinsiyet sütununu 0 ve 1 olarak dönüştürme işlemi için LabelEncoder sınıfını çağırdık
from sklearn.preprocessing import LabelEncoder
```

```python
# LabelEncoder sınıfını le değişkenine atıyoruz
le = LabelEncoder()

#Bu dönüşümü gender isminde yeni bir sütuna yazıyoruz
data['gender'] = le.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
data['Sex'].value_counts()
```

```python
# Species sütunundaki isimleri kısaltma amaçlı değiştirme işlemi
data['Species'] = data['Species'].replace('Adelie Penguin (Pygoscelis adeliae)','Adelie')
data['Species'] = data['Species'].replace('Gentoo penguin (Pygoscelis papua)','Gentoo')
data['Species'] = data['Species'].replace('Chinstrap penguin (Pygoscelis antarctica)','Chinstrap')
```

```python
# Sex sütunuyla işimiz kalmadı, sildik
data.drop(labels=['Sex'], axis=1, inplace=True)
```

```python
# Species sütunundaki değerlerin sayısını reset index metoduyla yeni bir dataframe'e dönüştürdük
species_count = data['Species'].value_counts().reset_index()
```

```python
species_count
```

```python
# Arkasına çizgi çektik
sns.set_theme(style="whitegrid")

# Tür sayısı için bir bar grafiği oluşturduk
sns.barplot(data=species_count, x='Species', y='count', color='purple')
```

```python
# Adelie türündeki değerlerin vücut ağırlığını gösterdik
data[data['Species'] == 'Adelie']['Body Mass (g)']
```

```python
# Species sütunundaki birbirinden farklı değerlerin(unique) vücut ağrılıklarını çizgi grafiğinde gösterdik. 
# Label komutuyla türlerin etiketini gösterdik. Fill komutuyla hacim verdik.
for spec in data['Species'].unique():
    sns.kdeplot(data[data['Species'] == spec]['Body Mass (g)'], label=spec, fill=True)
    plt.legend()
```

```python
# Sütunlardaki sayısal değerleri almak için belirli bir aralık oluşturduk. 
# Bir önceki grafikte yaptığımız işlemi bu sefer tüm sayısal sütunlar için yaptık. 
# plt legend birden fazla grafik olduğunda etiket koymak için gerekiyor.
for col in data.columns[8:-4]:
    for spec in data['Species'].unique():
        sns.kdeplot(data[data['Species'] == spec][col], label=spec, fill=True)
        plt.legend()
    plt.show()
```

```python
# Veri setindeki tüm sayısal değerler arasındaki ilişkileri göstermek için pairplot metodunu kullandık. 
# Hue komutu ile aralarındaki ilişkiyi tür bazlı gösterdik. diag_kind ile histogram grafiği istediğimizi belirttik.
sns.pairplot(data, hue='Species', size=3, diag_kind='hist')
```

```python
# Machine Learning

# get_dummies metoduyla Island sütununun değerlerini 0 ve 1 olarak dönüştürüyoruz. 
# Satırdaki doğru değer için 1 diğer değerler için 0 değerlerine dönüştürür.
# Bu metodu tüm Island değerleri için yapıyor. Yani üç değer için üç sütun varsa, iki sütunla sonucu anlayabileceğimiz için bir sütunu düşürüyoruz.
# dtype=int ile True False yerine 0 ve 1 yazmasını sağlıyoruz.
island = pd.get_dummies(data[['Island']], drop_first=True, dtype=int)
```

```python
# 0 ve 1 olarak dönüştürdüğümüz island değerini veri setimizie ekledik. 
# Drop ile önceki Island sütununu kaldırıyoruz
data2 = pd.concat([data, island], axis=1).drop(['Island'],axis=1)
```

```python
data2.head()
```

```python
# target olarak data2 dataframe'indeki Species sütununu aldık. 
# features olarak sayısal değerlerin olduğu sütunları seçtik.
target, features = data2.Species, data2[['Culmen Length (mm)',
       'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)','gender', 'Island_Dream','Island_Torgersen']]
```

```python
# StandardScaler sınıfını çağırdık. 
# Bu sınıf veri setindeki özelliklerin birbiriyle karşılaştırılabilmesini sağlar
# StandardScaler, özellikleri, her bir özelliğin ortalaması çıkarılarak ve standart sapması ile bölünerek standartlaştırır.
from sklearn.preprocessing import StandardScaler
```

```python
features.head()
```

```python
# Standardscaler sınıfından bir obje oluşturduk
scaler = StandardScaler()
```

```python
# features tablosundaki istediğimiz aralığı belirttik.
# scaler.fit ile veriyi hesapladık.
scaler.fit(features.iloc[:,:4])
```

```python
# transform ile veriyi dönüştürdük ve üzerine yazdık.
features.iloc[:,:4] = scaler.transform(features.iloc[:,:4])
```

```python
features.iloc[:,:4]
```

```python
# target değerlerini 0, 1 ve 2 şeklinded dönüştürdük.
target_encoded = le.fit_transform(target)
```

```python
target_encoded
```

```python
# Bu fonksiyon bir makine öğrenmesi modeli eğiteceğimiz zaman elimizdeki veri seti sınırlıysa ve gelecekte veri gelmeyecekse train ve test olarak ikiye ayırıyoruz.
from sklearn.model_selection import train_test_split
```

```python
# Train ve test değerlerimizi belirledik, test için gerekli veriyi %20 olarak belirledik. 
X_train, X_test, y_train, y_test = train_test_split(features, target_encoded, test_size=0.2, random_state=42)
```

```python
# Karar ağacı çağırdık. 
# Veri setindeki özellikleri ve hedef değişkeni kullanarak bir karar ağacı modeli oluşturur.
# Bu model yeni bir veriyi sınıflandırmak için kullanılır.
from sklearn.tree import DecisionTreeClassifier

# Modelin performansını ölçmek için çeşitli sınıflar çağırdık
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Sınıftan obje oluşturduk
tree = DecisionTreeClassifier()

# Karar ağacına x ve y değerlerimizi vererek model eğitiyoruz.
tree.fit(X_train, y_train)
```

```python
# Train verisiyle eğittiğimiz modeli predict metodu kullanarak test verileriyle tahmin etmeye çalışıyoruz.
predict = tree.predict(X_test)
```

```python
# Gerçek değerlerimizle(y_test) tahmin ettiğimiz değerlerimizle(predict) ne kadar benzer ona bakıyoruz.
# confusion_matrix ise sonuçları özetler. 
confusion_matrix(y_test, predict)
```

```python
# heatmap ile sonucu görselleştiriyoruz
sns.heatmap(confusion_matrix(y_test, predict),annot=True, cmap='coolwarm', cbar=False)
```

```python
# classification_report ile modelin doğruluk, hassasiyet, özgüllük ve F1-score gibi değerlerine bakkıyoruz.
print(classification_report(y_test, predict))
```

```python
# Doğruluk oranına yüzde cinsinden bakıyoruz
print('Accuracy Score = %',accuracy_score(y_test, predict) * 100)
```

```python
# Veri setinin son halini final data isimli yeni bir dataframe olarak kaydediyoruz.
final_data = pd.concat([features, pd.Series(target_encoded, name='Target')],axis=1)
```

```python
final_data
```

```python
# final data değişkenini 'penguins_scaled.csv' olarak csv formatına dönüştürdük.
# indexleme işlemini devre dışı bıraktık.
final_data.to_csv('penguins_scaled.csv',index=False, encoding='utf-8')
```
